```python
import os
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
#%pylab inline
import pylabsetup
import pylab as pl
%pylab inline


```

```python
np.random.seed(555)
d1 = np.random.randn(10)*1.2+5
d2 = np.random.randn(5)*2
```

```python
x=np.arange(-10,10,0.01)

gs = lambda x,mu,s: np.exp((-(x-mu)**2)/2/s/s)*1.0/np.sqrt(2*np.pi)/s

fig = pl.figure(5,figsize=(10,5))
fig.patch.set_facecolor('black')
ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')#

pl.scatter( d1, np.zeros(10), color='SteelBlue', s=300)
pl.scatter( d2, np.zeros(5), color='IndianRed', s=300)
pl.title("problem solved")
ax.yaxis.set_ticklabels("");
```

```python
x=np.arange(-15,15,0.01)
np.random.seed(555)
fig = pl.figure(5,figsize=(10,5))
fig.patch.set_facecolor('black')
dall=np.concatenate([d1,d2])
ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')#

pl.scatter( d2, np.zeros(5), color='SteelBlue', s=500)
pl.scatter( d1, np.zeros(10),  color='IndianRed', s=500)
pl.scatter( d1[(d1<4.5) & (d1>2)], np.zeros(10)[(d1<4.5) & (d1>2)], 
           color='SteelBlue', s=200)
pl.scatter( d1[(d1>5.5)], np.zeros(10)[(d1>5.5)], 
           color='SteelBlue', s=500)
pl.scatter( d1[(d1>4.5)], np.zeros(10)[(d1>4.5)], 
           color='SteelBlue', s=400)
pl.scatter( d2[(d2>2.5)], np.zeros(5)[(d2>2.5)], 
           color='IndianRed', s=200)
pl.scatter( d2[(d2<2.5)], np.zeros(5)[(d2<2.5)], 
           color='IndianRed', s=400)
pl.scatter( d2[(d2<0.5)], np.zeros(5)[(d2<0.5)], 
           color='IndianRed', s=500)

ax.plot(x,gs(x,min(dall),1),color='IndianRed',lw=3)
ax.plot(x,gs(x,max(dall),2),color='SteelBlue',lw=3)
ax.yaxis.set_ticklabels("");
```

```python
pd1 = gs(dall,min(dall), 1)
pd2 = gs(dall,max(dall), 1)
p1 = pd1/(pd1+pd2)
p2 = 1.0-p1

newm1 = np.average(dall, weights=p1)
newm2 = np.average(dall, weights=p2)
news1 = np.average((dall-newm1)**2, weights=p1)
news2 = np.average((dall-newm2)**2, weights=p2)
fig = pl.figure(5,figsize=(10,5))
fig.patch.set_facecolor('black')
dall=np.concatenate([d1,d2])
ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')#

pl.scatter( d2, np.zeros(5), color='SteelBlue', s=500)
pl.scatter( d1, np.zeros(10),  color='IndianRed', s=500)
pl.scatter( d1[(d1<4.5) & (d1>2)], np.zeros(10)[(d1<4.5) & (d1>2)], 
           color='SteelBlue', s=350)
pl.scatter( d1[(d1>5.5)], np.zeros(10)[(d1>5.5)], 
           color='SteelBlue', s=500)
pl.scatter( d1[(d1>4.5)], np.zeros(10)[(d1>4.5)], 
           color='SteelBlue', s=400)
pl.scatter( d2[(d2>2.5)], np.zeros(5)[(d2>2.5)], 
           color='IndianRed', s=100)
pl.scatter( d2[(d2<2.5)], np.zeros(5)[(d2<2.5)], 
           color='IndianRed', s=400)
pl.scatter( d2[(d2<0.5)], np.zeros(5)[(d2<0.5)], 
           color='IndianRed', s=500)
print newm1, np.sqrt(news1)
print newm2, np.sqrt(news2)

ax.plot(x,gs(x, newm1, np.sqrt(news1)),color='IndianRed',lw=3)
ax.plot(x,gs(x, newm2, np.sqrt(news2)),color='SteelBlue',lw=3)
```

```python
tol = 1e-9
oldm1 = -99
olds1 = -99
oldm2 = -99
olds2 = -99

while (np.abs(newm1-oldm1) + np.abs(newm2-oldm2) > tol or 
       np.abs(news1-olds1) + np.abs(news2-olds2) > tol):
    #print np.abs(newm1-oldm1) + np.abs(newm2-oldm2)
    #print np.abs(news1-olds1) + np.abs(news2-olds2)
    oldm1 = newm1
    oldm2 = newm2
    olds1 = news1
    olds2 = news2
    pd1 = gs(dall,newm1, np.sqrt(news1))
    pd2 = gs(dall,newm2, np.sqrt(news2))
    p1 = pd1/(pd1+pd2)
    p2 = 1.0-p1

    newm1 = np.average(dall, weights=p1)
    newm2 = np.average(dall, weights=p2)
    news1 = np.average((dall-newm1)**2, weights=p1)
    news2 = np.average((dall-newm2)**2, weights=p2)
    fig = pl.figure(figsize=(10,5))
    fig.patch.set_facecolor('black')
    dall=np.concatenate([d1,d2])
    ax = fig.add_subplot(111, axisbg='black')
    ax.tick_params(color='gray', labelcolor='gray')
    for spine in ax.spines.values():
        spine.set_edgecolor('gray')#

    pl.scatter( d1, np.zeros(10), color='SteelBlue', s=500)
    pl.scatter( d2, np.zeros(5),  color='IndianRed', s=500)
    print "mu & sigma of gaussian 1: ", newm1, np.sqrt(news1)
    print "mu & sigma of gaussian 2: ", newm2, np.sqrt(news2)

    ax.plot(x,gs(x, newm1, np.sqrt(news1)),color='IndianRed',lw=3)
    ax.plot(x,gs(x, newm2, np.sqrt(news2)),color='SteelBlue',lw=3)
```
